```python
import numpy as np
import matplotlib.pyplot as plt
#%plt.style.use("v_paper")
import pandas as pd
import deepdish as dd
```

```python
from pathlib import Path
```

```python
metadata_all = pd.read_excel(r"all_experiments_corrected_corrected0624.xlsx")
```

```python
folder = Path(r"Planewiselobesegmentation")
```

```python
metadata = []
flies = []
for i_row, flydata in metadata_all.iterrows():
    try:
        flies.append(dd.io.load(folder / (flydata.folder_name + ".h5")))
        metadata.append(flydata)
    except OSError:
        pass
    
metadata = pd.DataFrame(metadata)
```

```python
def calc_f0(trace):
    return np.mean(trace[:29])
```

```python
import cmocean as cmo
```

```python
cbase = cmo.tools.crop_by_percent(cmo.cm.curl,20, which="min")    
```

```python
cbase = cmo.tools.crop_by_percent(cbase, 20, which="max")    
```

```python
#cbase = cmo.cm.curl
```

```python
LobesGreek=['α1', 'α2', 'α3', "α'1", "α'2", "α'3", 'β1', 'β2', "β'1", "β'2",
       'γ1', 'γ2', 'γ3', 'γ4', 'γ5']
```

```python
Lobes=fly["lobe_traces_starts_ends"].keys()
Lobes
```

```python
dictionaryGreek = dict(zip(Lobes, LobesGreek))
```

```python
plt.close('all')
```

```python
fly = flies[16]
```

```python
if fly["metadata"]["stack_reversed"]:
    colors = cbase.reversed()
else:
    colors = cbase
n_lobes = len(fly["lobe_traces_starts_ends"])
fig, axes = plt.subplots(n_lobes, 1, gridspec_kw=dict(hspace=0.0), sharex=True, figsize=(8,10))
for i_lobe, (name, (i_st, i_en)) in enumerate(fly["lobe_traces_starts_ends"].items()):
    ax = axes[i_lobe]
    for i_trace in range(i_st, i_en):
    #for i_trace in range(200):
        trace = fly["traces_mn"][i_trace]
        f0 = calc_f0(trace)
        ax.plot((trace-f0)/f0, lw=0.5, color=colors((i_trace-i_st)/(i_en-i_st)))
    ax.text(0,0.5, "planes {}".format(fly["plane_ids"][i_st]), color=(colors(0.)))
    ax.text(10,0.5, "{}".format(fly["plane_ids"][i_en-1]-1), color=colors(1.))
    ax.axvline(30, color="r")
    ax.set_ylabel(dictionaryGreek[name])
fig.align_ylabels()
fig.text(0, 0, 'ΔF/F', va='center', rotation='vertical')

fig.suptitle(fly["metadata"]["folder_name"]+" "+("stack reversed" if fly["metadata"]["stack_reversed"] else ""))

fig.savefig('SinglePlaneTraces.svg')
fig.savefig('SinglePlaneTracesDir2.png')
```
