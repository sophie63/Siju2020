```python
data_tastequi=pd.read_excel("Taste_data.xlsx",'quinine').transpose()

data_tastesuc=pd.read_excel("Taste_data.xlsx",'sucrose').transpose()

db1=data_tastesuc.stack().reset_index()

db2=data_tastequi.stack().reset_index()

#taste_mean=pd.concat([suc_mean,qui_mean], ignore_index=True, sort=False)

db2['taste']='quinine'
db1['taste']='sucrose'

data_taste = pd.concat([db1, db2], ignore_index=True, sort=False)

data_taste.columns = ['lobe', 'num','df_f','taste']

data_taste['df_f_norm']=0
data_taste['norm']=0

Nums=[170,110,29,177,176,173,188,151,174,175,168,170]

for taste in ['quinine', 'sucrose']:
    for i in Nums:
        if any((data_taste['num']==i) & (data_taste['taste']==taste)):
            subData=data_taste[data_taste['num']==i][data_taste['taste']==taste]
            Norm=np.float(subData[subData['lobe']=='Mean'].df_f)
            data_taste.loc[(data_taste['num']==i) & (data_taste['taste']==taste),'norm']=Norm


data_taste.df_f_norm=data_taste.df_f-data_taste.norm
```

# Get individual p values

```python
ptaste=np.zeros(15)
i=0
for lobe in Lobes:
    lobe_datab=data_taste[data_taste['lobe']==lobe]
    sub_lobe_data_suc=lobe_datab[lobe_datab['taste']=='sucrose'].dropna()
    sub_lobe_data_qui=lobe_datab[lobe_datab['taste']=='quinine'].dropna()
    test=scipy.stats.mannwhitneyu(sub_lobe_data_suc["df_f_norm"], sub_lobe_data_qui["df_f_norm"], alternative='two-sided')
    ptaste[i]=test[1]
    i=i+1
    if i==14:
        EffectSize=(np.mean(sub_lobe_data_suc["df_f_norm"])-np.mean(sub_lobe_data_qui["df_f_norm"]))/np.sqrt((np.std(sub_lobe_data_suc["df_f_norm"])*np.std(sub_lobe_data_suc["df_f_norm"])+np.std(sub_lobe_data_qui["df_f_norm"])*np.std(sub_lobe_data_qui["df_f_norm"]))/2)
print(EffectSize)
print(ptaste)
```

```python
power_analysis = ssp.TTestIndPower()
power_analysis.solve_power(effect_size = EffectSize, 
                                         power = 0.8, 
                                         alpha = 0.05/60)
```

```python
Pcorrecttaste=sm.stats.multitest.multipletests(ptaste, alpha=0.05, method='simes-hochberg', is_sorted=False, returnsorted=False)
Pcorrecttaste
```

```python
pvinegar=np.zeros(15)
i=0
for lobe in Lobes:
    lobe_datab=lobe_data_w_vinegar[lobe_data_w_vinegar['lobe']==lobe]
    sub_lobe_data=lobe_datab[lobe_datab.hours_starved.isin({24,48})]
    sub_lobe_data_low=sub_lobe_data[sub_lobe_data['stimulus']=='Vinegar']
    sub_lobe_data_high=sub_lobe_data[sub_lobe_data['stimulus']=='high_Vinegar']
    test=sm.stats.weightstats.ttest_ind(sub_lobe_data_low["df_f"], sub_lobe_data_high["df_f"], alternative='two-sided', usevar='pooled', weights=(None, None), value=0)
    pvinegar[i]=test[1]
    i=i+1
```
