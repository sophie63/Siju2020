```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import deepdish as dd
from scipy.signal import savgol_filter

%matplotlib inline
```

```python
import statsmodels.formula.api as smf

from statsmodels.stats.anova import anova_lm

import statsmodels.api as smapi


import seaborn as sns

import statsmodels as sm
```

```python
esum = dd.io.load("experiments_summary.h5")

lobe_data = pd.DataFrame(esum["lobe_summary"])

lobe_activities = np.row_stack(esum["df_fs"])

lobe_data["df_f"] = np.mean(lobe_activities[:, 31:35], 1)
```

```python
lobe_data["adaptnorm"] = -lobe_data.adaptation/lobe_data.df_mean
```

```python
plt.plot(lobe_activities.T)
plt.show()
```

```python
valence_map = {
"Vinegar": 1,
"Yeast": 1,
"Isoamylacetate": np.nan,
"2-Heptanone": np.nan,
"1-Heptanone": np.nan,
    
"cVA": np.nan,
        "3Octanol": np.nan,
        "Ethanol": np.nan,

"1-Hexanol": np.nan,
"Peppermint": -1,
"4MCH": np.nan,
"Citronella": -1,
"Geosmin": -1,
}

lobe_data["valence"] = [valence_map[s] for s in lobe_data.stimulus]
```

```python
lobes = lobe_data.lobe.unique()
n_lobes = len(lobes)
n_time = lobe_activities.shape[1]
```

```python
Lobes=lobe_data.lobe.unique()
```

```python
Stimuli=lobe_data.stimulus.unique()
```

```python
len(Stimuli)
```

# All variables

```python
res4 = smf.ols(formula="df_f ~ C(lobe)*(stimulus+order_presented+ C(dorsal_to_ventral))+(C(lobe)-1):hours_starved- 1", data=lobe_data).fit()

res4.summary()
```

```python
aov_table = anova_lm(res4, typ=2)
print(res4.rsquared)
aov_table
```

```python
aov_table.to_csv("FullAnovaTable.csv")
```

```python
import statsmodels.api as smapi

residuals = res4.resid 
fig = smapi.qqplot(residuals, line='s')
plt.show()
```

```python
resval = smf.ols(formula="df_f ~ C(lobe)*(order_presented+ C(dorsal_to_ventral))+C(lobe):valence+C(lobe):hours_starved", data=lobe_data).fit()

resval.summary()
```

```python
aov_table = anova_lm(resval, typ=2)
print(resval.rsquared)
aov_table
```

```python
aov_table.to_csv("AnovaTableValence.csv")
```

```python
res4 = smf.ols(formula="df_f ~ C(lobe)*(order_presented+ C(dorsal_to_ventral))+(C(lobe)-1):hours_starved- 1", data=lobe_data).fit()

aov_table = anova_lm(res4, typ=2)
print(res4.rsquared)
aov_table
```

```python
%%R
library(lme4)
m_rt <- lmer(df_f ~ 1+lobe*valence+lobe*(stimulus|valence),
        data = lobe_data)

rfs <- ranef(m_rt)$cat
ffs <- fixef(m_rt) 
```

```python
%%R
library(lme4)
Summixedmod=summary(m_rt)
anovamixed=anova(m_rt)
capture.output(Summixedmod, file = "SummaryModel.txt")
capture.output(anovamixed, file = "anova results.txt")
```

```python
#Hunger
print(lobes)
LobeValues=np.array(res4.params[len(res4.params)-15:len(res4.params)]).reshape(1,15)
plt.imshow(LobeValues)
plt.colorbar()
plt.savefig('ColorHungerBar.svg')
```

```python
Vmax=max(abs(LobeValues.T))
```

```python
fig, axs = plt.subplots(5, 1, figsize=(10, 3))
i=0
for ax, interp in zip(axs, [3,6,8,10,15]):
    OneLobe=LobeValues[0,range(i,interp)].reshape(1,interp-i)
    ax.imshow(OneLobe, vmin=-Vmax,vmax=Vmax)
    ax.grid(False)
    #ax.axis('off')
    ax.xaxis.set_ticks_position('none') 
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.yaxis.set_ticks_position('none')  
    i=interp

plt.show()
fig.savefig('ColorHunger.svg')
```
