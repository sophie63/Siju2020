```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#plt.style.use("v_paper")
import deepdish as dd
from scipy.signal import savgol_filter

%matplotlib inline
```

```python
esum = dd.io.load("experiments_summary.h5")

lobe_data = pd.DataFrame(esum["lobe_summary"])

lobe_activities = np.row_stack(esum["df_fs"])
lobe_data["df_f"] = np.mean(lobe_activities[:, 31:35], 1)
lobe_data["Î”F/F"] = np.mean(lobe_activities[:, 31:35], 1)
```

```python
plt.plot(lobe_activities.T)
plt.show()
```

Other pre-requisites

```python
lobe_data["hours starved"]=lobe_data["hours_starved"]
```

```python
valence_map = {
"Vinegar": '1',
"Yeast": '1',
"Isoamylacetate": np.nan,
"2-Heptanone": np.nan,
"1-Heptanone": np.nan,
    
"cVA": np.nan,
        "3Octanol": np.nan,
        "Ethanol": np.nan,

"1-Hexanol": np.nan,
"Peppermint": '-1',
"4MCH": np.nan,
"Citronella": '-1',
"Geosmin": '-1',
}

lobe_data["valence"] = [valence_map[s] for s in lobe_data.stimulus]
```

```python
Lobes=lobe_data.lobe.unique()
Lobes
```

```python
LobesGreek=['Î±1', 'Î±2', 'Î±3', "Î±'1", "Î±'2", "Î±'3", 'Î²1', 'Î²2', "Î²'1", "Î²'2",
       'Î³1', 'Î³2', 'Î³3', 'Î³4', 'Î³5']
```

```python
dictionaryGreek = dict(zip(Lobes, LobesGreek))
```

```python
lobe_data["greek_lobe"] = [dictionaryGreek[l] for l in lobe_data.lobe]
```

# Plot all the data

Peak response

```python
sns.set_style("whitegrid",{"axes.edgecolor": "0",'grid.color': '0'})
sns.set_context("notebook",font_scale=2)
plt.figure(figsize=(16, 6))
b = sns.boxplot(x="greek_lobe", y="Î”F/F",data=lobe_data)

b.set(xlabel='Lobe')

for i,box in enumerate(b.artists):
    box.set_edgecolor('black')
    box.set_facecolor('white')

    # iterate over whiskers and median lines
    for j in range(6*i,6*(i+1)):
         b.lines[j].set_color('black')
plt.yticks([0,1,2,3])  
plt.ylim([-0.5,4.1])
plt.savefig('AllLolesDFF.svg')
    
plt.show()
```

```python
sns.set_style("whitegrid",{"axes.edgecolor": "0",'grid.color': '0'})
sns.set_context("notebook",font_scale=2)
plt.figure(figsize=(16, 6))
b = sns.boxplot(x="greek_lobe", y="pre_std",data=lobe_data)
b.set(xlabel='Lobe',ylabel='Standard deviation')

for i,box in enumerate(b.artists):
    box.set_edgecolor('black')
    box.set_facecolor('white')

    # iterate over whiskers and median lines
    for j in range(6*i,6*(i+1)):
         b.lines[j].set_color('black')

plt.ylim([0,0.4])
plt.savefig('Lobesprestd.svg')
plt.show()
```

```python
lobe_data["adaptnorm"] = -lobe_data.adaptation/lobe_data.df_mean

lobe_data.adaptnorm[lobe_data.df_mean<0.3]=np.nan
```

```python
lobe_data["adaptnormperc"]=lobe_data.adaptnorm*100
```

```python
sns.set_style("whitegrid",{"axes.edgecolor": "0",'grid.color': '0'})
sns.set_context("notebook",font_scale=2)
plt.figure(figsize=(16, 6))
b = sns.boxplot(x="greek_lobe", y="adaptnormperc",data=lobe_data)
b.set(xlabel='Lobe',ylabel='% Adaptation/odor presentation')

for i,box in enumerate(b.artists):
    box.set_edgecolor('black')
    box.set_facecolor('white')

    # iterate over whiskers and median lines
    for j in range(6*i,6*(i+1)):
         b.lines[j].set_color('black')
b.xaxis.set_ticks_position('top')
b.xaxis.set_label_position('top')
b.set(ylim=(-120, 100))
plt.yticks([-50, 0,50])
plt.savefig('Adaptation.svg')            
plt.show()
```
